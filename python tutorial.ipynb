```python
## inheritance: a class inheriting features from the super class
## parent class
class animal:
    def sound(self):
        print("animals makes sound")     ## super class
class dog(animal):
    def bark(self):
        print("dog barks")
object1=dog()                              ## sub class

object1.sound()
object1.bark()
```

```python
## inheritance using init constructor
class vechile:
    def __init__(self,milegae,cost):
        self.milegae=milegae
        self.cost=cost
    def show(self):                               ## super  class
        print("milegae",self.milegae)
        print("cost",self.cost)

class car(vechile):
      color="blue"
object1=car(500,500)
object1.show()
print(object1.color)
      
    
```

```python
## types of inheritance
##(1) single inheritance
##(2) multiple inheritance
##(3) multi level 
##(4) hybrid
```

```python
##(1) single inheritance : having one parent class and one child class
# Create a parent class Person:

# attributes: name, age

# method: display details

# 👉 Create child class Student:

# add attribute: marks

# print all details


class parent:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print("name",self.name)
        print("age",self.age)
class child(parent):
    def __init__(self,name,age,marks):
        super().__init__(name,age)
        self.marks=marks

    def show(self):
        super().show()
        print("marks",self.marks)

object1=child("keerthi",21,95)
object1.show()

    
    
    
```

```python
##(2) another exmaple for single inhritance
class parent:
    def show_parent(self):
        print("iam the parent")
class child(parent):
    def show_child(self):
        print("iam the child")
object1=child()
object1.show_parent()
object1.show_child()
```

```python
##(3) multiple inheritance: more than one parent
class father:
    def show_father(self):
        print("iam the father")
class mother:
    def show_mother(self):
        print("iam the mother")
class child(father,mother):
    def show_child(self):
        print("iam the child")
object1=child()
object1.show_father()
object1.show_mother()
object1.show_child()
        
    
```

```python
## another example

## parent class
class parent1:
    def assign_string_one(self,string1):
        self.string1=string1
    def show_string_one(self):
        print("string_one",self.string1)
## parent class 2
class parent2:
    def assign_string_two(self,string2):
        self.string2=string2
    def show_string_two(self):
        print("string_two",self.string2)

## child class
class child(parent1,parent2):
    def assign_string_three(self,string3):
        self.string3=string3
    def show_string_three(self):
        print("string_three",self.string3)
## creating the object
object1=child()
object1.assign_string_one("one")
object1.assign_string_two("two")
object1.assign_string_three("three")
object1.show_string_one()
object1.show_string_two()
object1.show_string_three()
```

```python
## multi level inheritance: inheriting from a class that is already inherits from another class
## parent
## child
## grand child

## parent class
class parent:
    def assign_name(self,name):
        self.name=name
    def show_name(self):
        print(self.name)
## creating a child class
class child(parent):
    def assign_age(self,age):
        self.age=age
    def show_age(self):
        print(self.age)
## creating the grand child class
class grandchild(child):
    def assign_gender(self,gender):
        self.gender=gender
    def show_gender(self):
        print(self.gender)


object1=grandchild()
object1.assign_name("keerthi")
object1.assign_age(20)
object1.assign_gender("female")
object1.show_name()
object1.show_age()
object1.show_gender()
```
